```python
df_ctry = pd.read_excel('Country-Code.xlsx')
df_ctry.head()
```

```python
final_df = pd.merge(df,df_ctry, on='Country Code', how='left')
final_df.head()
```

```python
final_df.columns
```

```python
ctry_names = final_df.Country.value_counts().index
```

```python
ctry_values = final_df.Country.value_counts().values
```

```python
# Top 3 countries using zomato
plt.pie(ctry_values[:3], labels = ctry_names[:3], autopct='%1.2f%%')
```

Observation: Zomato maximum transactions are from India <br>
Next comes, USA and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations

1. Whenever the rating is between 4.5 - 4.9 - Food delivery is Excellent! <br>
2. When rating is between 4.0 - 4.4 - Rating is very good <br>
3. When rating is between 3.5 - 3.9 - Rating is good and so on

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
#Barplot
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', palette=['black', 'red', 'orange', 'yellow', 'green', 'green'] ,data = ratings)
```

## Observations

1. Not rated count is very high <br>
2. Maximum number of rating between 2.6 -  4.5

```python
#Countplot
sns.countplot(x = final_df['Rating color'], data = ratings, palette=['black', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
# Find the country names that has given 0 rating
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

## Observations
1. India has the highest number of non-rating Zomato orders <br>
2. UK has the lowest count


```python
# Find which currency is used by which country
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# Find which countries have online deliveries
final_df[final_df['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index().rename(columns={0:'Count'})
```
