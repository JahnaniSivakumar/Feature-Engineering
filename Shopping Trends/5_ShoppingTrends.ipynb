## Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Reading the file

```python
dataset = pd.read_csv('shopping_trends.csv')
```

## Exploring the dataset

```python
# Displaying the first 5 records
dataset.head()
```

## Observations
Customers purchase different category items from different locations. We can also see their gender, age, preferred purchase method, etc,.


```python
# Summarizing the dataset
dataset.info()
```

## Observation
There are 18 attributes in the dataset, out of which we have 5 numeric features and there is no null value in any feature.

```python
#Numerical summary
dataset.describe()
```

## Observation
There are a total of 3,900 records, with customers aged 18 and above making purchases. On average, each customer spends $59.76.

```python
# checking the total number of records
dataset.shape
```

```python
# checking if there is any null values
dataset.isna().sum()
```

```python
# checking if there is any duplicate values
dataset.duplicated().sum()
```

```python
# Age Distribution
dataset['Age'].plot(kind='hist')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price Distribution
dataset['Purchase Amount (USD)'].plot(kind='hist')
plt.title('Purchase Amount (USD)')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset.columns
```

```python
# Gender Distribution
dataset['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

## Observation
We expected female shoppers to lead the way, but guess what? More male customers are shopping

```python
# review rating distribution across gender
dataset.boxplot(column='Review Rating', by='Gender')
plt.title('Review Rating Vs Gender')
plt.xlabel('Gender')
plt.ylabel('Review Rating')
plt.show()
```

```python
# Top 2 categories with high profit
dataset.groupby('Category')['Purchase Amount (USD)'].mean().sort_values(ascending=False).head(2)
```

```python
# Sort dataframe based on the review rating in the descending
dataset.sort_values('Review Rating', ascending=False, inplace=True)
```

```python
dataset
```

```python
# Subsribers and non-subscribers
dataset['Subscription Status'].value_counts().plot(kind = 'pie')
plt.title('Subscription Status')
plt.show()
```

```python
# Top 5 colors purchased by customers
dataset['Color'].value_counts().sort_values(ascending=False).head(5)
```

```python
# Top season with highest profit
dataset.groupby('Season')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```

```python
# Purchasing amount sum across seasons
dataset.groupby('Season')['Purchase Amount (USD)'].sum().plot(kind = 'pie')
plt.title('Season')
plt.show()
```

```python
# Review Rating across seasons
dataset.groupby('Season')['Review Rating'].mean()
```

```python
# Purchase amount distribution in Frequency of Purchase
dataset.boxplot(column='Purchase Amount (USD)', by='Frequency of Purchases')
#plt.title('Purchase Amount vs Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Mostly used preferred payment method
dataset['Preferred Payment Method'].value_counts().sort_values(ascending=False)
```

```python
# sum of purchase across payment methods
dataset.groupby('Preferred Payment Method')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```

```python
pd.pivot_table(dataset, values='Purchase Amount (USD)', index= 'Location', columns= 'Item Purchased', aggfunc= 'mean').head()
```

```python
# Top selling size
dataset.groupby('Size')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
# what happens to review rating based on discounts
dataset.groupby('Discount Applied')['Review Rating'].mean()
```

```python
dataset.groupby('Color')['Category'].value_counts()
```

```python
dataset.groupby('Frequency of Purchases')['Previous Purchases'].median()
```

```python
# Distribution of purchases across categories
dataset['Category'].value_counts().plot(kind='pie')
plt.title('Distribution of purchases by category')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# review rating across colors

dataset.groupby('Color')['Review Rating'].mean().sort_values(ascending=False).head(5).plot(kind='bar')
plt.title('Review Rating Vs Color')
plt.xlabel('Color')
plt.ylabel('Review Rating')
plt.show()
```

## Observations

🔍While analyzing customer purchase behavior, shoppers aged 18 and above drive sales, with an average spend of $59.76 per customer.<br>
🔍We expected female shoppers to lead the way, but guess what? More male customers are shopping - the ones giving the highest ratings!(“The Aha moment”)<br>
🔍Footwear and clothing lead the way, with the highest average purchase amounts topping the charts.<br>
🔍The top 5 colors flying off the shelves are - Olive, Yellow, Silver, Teal, and Green.<br>
🔍While PayPal is the go-to choice for many, the credit card method takes the lead in total spending.
